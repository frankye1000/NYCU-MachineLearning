```python
from PIL import Image
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib
from scipy.spatial.distance import cdist
```

## Image Processing

```python
def read_image(path):
    imgs = os.listdir(path)
    n = len(imgs)
    
    W = len(np.asarray(Image.open(os.path.join(path, imgs[0]))))
    H = len(np.asarray(Image.open(os.path.join(path, imgs[0])))[0])
    
    images = np.zeros((W*H, n))
    labels = np.zeros(n).astype('uint8')
    
    for i in range(n):
        labels[i] = int(imgs[i][7:9]) - 1
        img = np.asarray(Image.open(os.path.join(path, imgs[i])).resize((W,H),Image.ANTIALIAS)).flatten()
        images[:, i] = img
        
    return images, labels, H, W

def show_Eigenface(U, H, W):
    #matplotlib.use("Qt5Agg")
    for i in range(25):
        plt.subplot(5, 5, i+1)
        plt.imshow(U[:, i].reshape(H,W), cmap='gray')
        
    
    #plt.get_current_fig_manager().window.showMaximized()
    
    plt.show()
    
def show_reconstruction(X, X_recon, H, W):
    randimage = np.random.choice(X.shape[1], 10)
    #matplotlib.use("Qt5Agg")
    for i in range(10):
        # show random 10 origin image
        plt.subplot(2, 10, i+1)
        plt.imshow(X[:, randimage[i]].reshape(H,W), cmap='gray')
        # show reconstruction of above 10 image
        plt.subplot(2, 10, i+11)
        plt.imshow(X_recon[:, randimage[i]].reshape(H,W), cmap='gray')
        
    
    #plt.get_current_fig_manager().window.showMaximized()
    plt.show()
```

## Kernel

```python
def rbfkernel(X1, X2, gamma):
    return np.exp(-gamma*cdist(X1, X2, 'sqeuclidean'))

def polykernel(X):
    K = np.zeros(shape=(len(X), len(X)))
    for i in range(len(X)):
        for j in range(len(X)):
            k = 1 + np.dot(X[i].T,X[j]) ## K(i,j) = ( 1 + x(i).T . x(j) )^p
            k = k**2
            K[i][j] = k
    return K
```

## Face Recognition

```python
def face_recognition(X_test, y_test, Z, y, U, X_mean, k):
    # diemnsion reduction (projection)
    Z_test=U.T@(X_test-X_mean)

    # knn
    pred = np.zeros(Z_test.shape[1])
    for i in range(Z_test.shape[1]):
        dist = np.zeros(Z.shape[1])
        for j in range(Z.shape[1]):
            dist[j] = np.sum(np.square(Z_test[:,i] - Z[:,j]))
        sort_index=np.argsort(dist)
        nn = y[sort_index[:k]]
        unique, counts = np.unique(nn, return_counts=True)
        nn = [k for k,v in sorted(dict(zip(unique, counts)).items(), key=lambda item: -item[1])]
        pred[i] = nn[0]

    acc = np.count_nonzero((y_test - pred) == 0) / len(y_test)
    print(f'accuracy: {acc*100:.2f}%')
```

# Eigenface with PCA

```python
def PCA(X, dim=None):
    # centering matrix
    X_mean = np.mean(X, axis=1).reshape(-1,1)
    X_center = X - X_mean
    cov_matrix = X_center.T@X_center
    
    # PCA
    eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
    eigen_sort_index = np.argsort(-eigenvalues)
    if dim is None:
        for eigenvalue, i in zip(eigenvalues[eigen_sort_index], np.arange(len(eigenvalues))):
            # reserve the eigen value > 0
            if eigenvalue <= 0:
                eigen_sort_index = eigen_sort_index[:i]
                break
    else:
        eigen_sort_index = eigen_sort_index[:dim]
        
    eigenvalues = eigenvalues[eigen_sort_index]
    
    # from X.T@X eigenvector to X@X.T eigenvector
    eigenvectors = X_center@eigenvectors[:, eigen_sort_index]
    eigenvectors_norm = np.linalg.norm(eigenvectors,axis=0)
    eigenvectors = eigenvectors/eigenvectors_norm
    
    return eigenvalues, eigenvectors, X_mean

def main_PCA():
    train_path = 'Yale_Face_Database\\Training'
    test_path = 'Yale_Face_Database\\Testing'
    
    # read the pmg file and flateen the image and get the label by file name
    X, y, H, W = read_image(train_path)
    X_test, y_test, H_test, W_test = read_image(test_path)
    
    
    eigenvalues, eigenvectors, X_mean = PCA(X)
    U = eigenvectors.copy()
    
    # Part 1. (a) show first 25 Eigenface
    show_Eigenface(U, H, W)
    
    # Part 1. (b) show 10 reconstruction Eigenface
    ## diemnsion reduction (projection)
    Z = U.T@(X-X_mean)

    ## reconstruction
    X_recon = U@Z + X_mean
    show_reconstruction(X, X_recon, H, W)
    
    # Part 2. face recognition
    face_recognition(X_test, y_test, Z, y, U, X_mean, 5)
    
```

```python
main_PCA()
```

# Eigenface with KPCA

```python
def KPCA(X, dim, kernel):
    if kernel == 'RBF':
        gamma = 0.01
        K = rbfkernel(X, X, gamma)
        
    elif kernel == 'polynomial':
        K = polykernel(X)

    # centering matrix
    K_mean = np.mean(K, asis=0)
    K_center = K - K_mean
        
    # PCA
    eigenvalues, eigenvectors = np.linalg.eig(K_center)
    eigen_sort_index = np.argsort(-eigenvalues)

    if dim is None:
        for eigenvalue, i in zip(eigenvalues[eigen_sort_index], np.arange(len(eigenvalues))):
            # reserve the eigen value > 0
            if eigenvalue <= 0:
                eigen_sort_index = eigen_sort_index[:i]
                break
    else:
        eigen_sort_index = eigen_sort_index[:dim]
    
    eigenvalues = eigenvalues[eigen_sort_index]
    
    # from X.T@X eigenvector to X@X.T eigenvector
    eigenvectors = X_center@eigenvectors[:, eigen_sort_index]
    eigenvectors_norm = np.linalg.norm(eigenvectors,axis=0)
    eigenvectors = eigenvectors/eigenvectors_norm
    
    return eigenvalues, eigenvectors, X_mean
    
    
def main_KPCA():
    train_path = 'Yale_Face_Database\\Training'
    test_path = 'Yale_Face_Database\\Testing'
    
    # read the pmg file and flateen the image and get the label by file name
    X, y, H, W = read_image(train_path)
    
    # Part 3. (a) RBF kernel PCA
    print('RBF kernel PCA')
    eigenvalues, eigenvectors, X_mean = KPCA(X, None, 'RBF')
    U = eigenvectors.copy()
    ## diemnsion reduction (projection)
    Z = U.T@(X-X_mean)
    face_recognition(X_test, y_test, Z, y, U, X_mean, 3)
    
    # Part 3. (b) polynomial kernel PCA
    print('polynomial kernel PCA')
    eigenvalues, eigenvectors, X_mean = KPCA(X, None, 'polynomial')
    U = eigenvectors.copy()
    ## diemnsion reduction (projection)
    Z = U.T@(X-X_mean)
    face_recognition(X_test, y_test, Z, y, U, X_mean, 3)
```

```python
main_KPCA()
```

# Fisherface with LDA

```python
def LDA(X, y):
    n = X.shape[0]
    X_mean = np.mean(X, axis=1).reshape(-1, 1)
    
    classes_mean = np.zeros((n, 15))  # 15 classes
    for i in range(X.shape[1]):
        classes_mean[:, y[i]] += X[:, y[i]]
    classes_mean /= 9
    
    # within-class scatter
    S_within = np.zeros((n, n))
    for i in range(X.shape[1]):
        d = X[:, y[i]].reshape(-1,1) - classes_mean[:, y[i]].reshape(-1,1)
        S_within += d @ d.T

    # between-class scatter
    S_between = np.zeros((n, n))
    for i in range(15):
        d = classes_mean[:, i].reshape(-1,1) - X_mean
        S_between += 9 * d @ d.T

    eigenvalues, eigenvectors = np.linalg.eig(np.linalg.inv(S_within)@S_between)
    sort_index = np.argsort(-eigenvalues)
    sort_index = sort_index[:-1]  # reduce 1 dim

    eigenvalues = np.asarray(eigenvalues[sort_index].real, dtype='float')
    eigenvectors = np.asarray(eigenvectors[:, sort_index].real, dtype='float')

    return eigenvalues,eigenvectors
    
def main_LDA():
    train_path = 'Yale_Face_Database\\Training'
    test_path = 'Yale_Face_Database\\Testing'
    
    # read the pmg file and flateen the image and get the label by file name
    X, y, H, W = read_image(train_path)
    X_test, y_test, H_test, W_test = read_image(test_path)
    
    
    eigenvalues_pca, eigenvectors_pca, X_mean = PCA(X, 31)
    X_pca = eigenvectors_pca.T@(X-X_mean)
    eigenvalues_lda, eigenvectors_lda = LDA(X_pca, y)

    # Transform matrix
    U = eigenvectors_pca@eigenvectors_lda

    # Part 1. (a) show first 25 Fisherface
    show_Eigenface(U, H, W)
    
    # Part 1. (b) show 10 reconstruction Fisherface
    ## diemnsion reduction (projection)
    Z = U.T@(X-X_mean)

    ## reconstruction
    X_recon = U@Z + X_mean
    show_reconstruction(X, X_recon, H, W)
    
    # Part 2. face recognition
    face_recognition(X_test, y_test, Z, y, U, X_mean, )
```

```python
main_LDA()
```

# Fisherface with KLDA

```python
def KLDA(X, y, kernel):
    if kernel == 'RBF':
        gamma = 0.01
        K = rbfkernel(X, X, gamma)
        
    elif kernel == 'polynomial':
        K = polykernel(X)
        
        
    n = K.shape[0]
    K_mean = np.mean(K, axis=1).reshape(-1, 1)
    
    classes_mean = np.zeros((n, 15))  # 15 classes
    for i in range(K.shape[1]):
        classes_mean[:, y[i]] += K[:, y[i]]
    classes_mean /= 9
    
    # within-class scatter
    S_within = np.zeros((n, n))
    for i in range(K.shape[1]):
        d = K[:, y[i]].reshape(-1,1) - classes_mean[:, y[i]].reshape(-1,1)
        S_within += d @ d.T

    # between-class scatter
    S_between = np.zeros((n, n))
    for i in range(15):
        d = classes_mean[:, i].reshape(-1,1) - K_mean
        S_between += 9 * d @ d.T

    eigenvalues, eigenvectors = np.linalg.eig(np.linalg.inv(S_within)@S_between)
    sort_index = np.argsort(-eigenvalues)
    sort_index = sort_index[:-1]  # reduce 1 dim

    eigenvalues = np.asarray(eigenvalues[sort_index].real, dtype='float')
    eigenvectors = np.asarray(eigenvectors[:, sort_index].real, dtype='float')

    return eigenvalues,eigenvectors

def main_KLDA():
    train_path = 'Yale_Face_Database\\Training'
    test_path = 'Yale_Face_Database\\Testing'
    
    # read the pmg file and flateen the image and get the label by file name
    X, y, H, W = read_image(train_path)
    X_test, y_test, H_test, W_test = read_image(test_path)
    
    # Part 3. (a) RBF kernel LDA
    eigenvalues_pca, eigenvectors_pca, X_mean = PCA(X, 31, 'RBF')
    X_pca = eigenvectors_pca.T@(X-X_mean)
    eigenvalues_lda, eigenvectors_lda = KLDA(X_pca, y)

    # Transform matrix
    U = eigenvectors_pca@eigenvectors_lda

    face_recognition(X_test, y_test, Z, y, U, X_mean, 3)
    
    # Part 3. (b) polynomial kernel LDA
    eigenvalues_pca, eigenvectors_pca, X_mean = PCA(X, 31, 'polynomial')
    X_pca = eigenvectors_pca.T@(X-X_mean)
    eigenvalues_lda, eigenvectors_lda = KLDA(X_pca, y)

    # Transform matrix
    U = eigenvectors_pca@eigenvectors_lda

    face_recognition(X_test, y_test, Z, y, U, X_mean, 3)
```

# t-sne

```python
#
#  tsne.py
#
# Implementation of t-SNE in Python. The implementation was tested on Python
# 2.7.10, and it requires a working installation of NumPy. The implementation
# comes with an example on the MNIST dataset. In order to plot the
# results of this example, a working installation of matplotlib is required.
#
# The example can be run by executing: `ipython tsne.py`
#
#
#  Created by Laurens van der Maaten on 20-12-08.
#  Copyright (c) 2008 Tilburg University. All rights reserved.

import numpy as np
import pylab


def Hbeta(D=np.array([]), beta=1.0):
    """
        Compute the perplexity and the P-row for a specific value of the
        precision of a Gaussian distribution.
    """

    # Compute P-row and corresponding perplexity
    P = np.exp(-D.copy() * beta)
    sumP = sum(P)
    H = np.log(sumP) + beta * np.sum(D * P) / sumP
    P = P / sumP
    return H, P


def x2p(X=np.array([]), tol=1e-5, perplexity=30.0):
    """
        Performs a binary search to get P-values in such a way that each
        conditional Gaussian has the same perplexity.
    """

    # Initialize some variables
    print("Computing pairwise distances...")
    (n, d) = X.shape
    sum_X = np.sum(np.square(X), 1)
    D = np.add(np.add(-2 * np.dot(X, X.T), sum_X).T, sum_X)
    P = np.zeros((n, n))
    beta = np.ones((n, 1))
    logU = np.log(perplexity)

    # Loop over all datapoints
    for i in range(n):

        # Print progress
        if i % 500 == 0:
            print("Computing P-values for point %d of %d..." % (i, n))

        # Compute the Gaussian kernel and entropy for the current precision
        betamin = -np.inf
        betamax = np.inf
        Di = D[i, np.concatenate((np.r_[0:i], np.r_[i+1:n]))]
        (H, thisP) = Hbeta(Di, beta[i])

        # Evaluate whether the perplexity is within tolerance
        Hdiff = H - logU
        tries = 0
        while np.abs(Hdiff) > tol and tries < 50:

            # If not, increase or decrease precision
            if Hdiff > 0:
                betamin = beta[i].copy()
                if betamax == np.inf or betamax == -np.inf:
                    beta[i] = beta[i] * 2.
                else:
                    beta[i] = (beta[i] + betamax) / 2.
            else:
                betamax = beta[i].copy()
                if betamin == np.inf or betamin == -np.inf:
                    beta[i] = beta[i] / 2.
                else:
                    beta[i] = (beta[i] + betamin) / 2.

            # Recompute the values
            (H, thisP) = Hbeta(Di, beta[i])
            Hdiff = H - logU
            tries += 1

        # Set the final row of P
        P[i, np.concatenate((np.r_[0:i], np.r_[i+1:n]))] = thisP

    # Return final P-matrix
    print("Mean value of sigma: %f" % np.mean(np.sqrt(1 / beta)))
    return P


def pca(X=np.array([]), no_dims=50):
    """
        Runs PCA on the NxD array X in order to reduce its dimensionality to
        no_dims dimensions.
    """

    print("Preprocessing the data using PCA...")
    (n, d) = X.shape
    X = X - np.tile(np.mean(X, 0), (n, 1))
    (l, M) = np.linalg.eig(np.dot(X.T, X))
    Y = np.dot(X, M[:, 0:no_dims])
    return Y


def tsne(X=np.array([]), no_dims=2, initial_dims=50, perplexity=30.0):
    """
        Runs t-SNE on the dataset in the NxD array X to reduce its
        dimensionality to no_dims dimensions. The syntaxis of the function is
        `Y = tsne.tsne(X, no_dims, perplexity), where X is an NxD NumPy array.
    """

    # Check inputs
    if isinstance(no_dims, float):
        print("Error: array X should have type float.")
        return -1
    if round(no_dims) != no_dims:
        print("Error: number of dimensions should be an integer.")
        return -1

    # Initialize variables
    X = pca(X, initial_dims).real
    (n, d) = X.shape
    max_iter = 1000
    initial_momentum = 0.5
    final_momentum = 0.8
    eta = 500
    min_gain = 0.01
    Y = np.random.randn(n, no_dims)
    dY = np.zeros((n, no_dims))
    iY = np.zeros((n, no_dims))
    gains = np.ones((n, no_dims))

    # Compute P-values
    P = x2p(X, 1e-5, perplexity)
    P = P + np.transpose(P)
    P = P / np.sum(P)
    P = P * 4.									# early exaggeration
    P = np.maximum(P, 1e-12)

    # Run iterations
    for iter in range(max_iter):

        # Compute pairwise affinities
        sum_Y = np.sum(np.square(Y), 1)
        num = -2. * np.dot(Y, Y.T)
        num = 1. / (1. + np.add(np.add(num, sum_Y).T, sum_Y))
        num[range(n), range(n)] = 0.
        Q = num / np.sum(num)
        Q = np.maximum(Q, 1e-12)

        # Compute gradient
        PQ = P - Q
        for i in range(n):
            dY[i, :] = np.sum(np.tile(PQ[:, i] * num[:, i], (no_dims, 1)).T * (Y[i, :] - Y), 0)

        # Perform the update
        if iter < 20:
            momentum = initial_momentum
        else:
            momentum = final_momentum
        gains = (gains + 0.2) * ((dY > 0.) != (iY > 0.)) + \
                (gains * 0.8) * ((dY > 0.) == (iY > 0.))
        gains[gains < min_gain] = min_gain
        iY = momentum * iY - eta * (gains * dY)
        Y = Y + iY
        Y = Y - np.tile(np.mean(Y, 0), (n, 1))

        # Compute current value of cost function
        if (iter + 1) % 10 == 0:
            C = np.sum(P * np.log(P / Q))
            print("Iteration %d: error is %f" % (iter + 1, C))

        # Stop lying about P-values
        if iter == 100:
            P = P / 4.
    
    # save scatter plot
        if iter % 10 == 0:
            pylab.clf()
            pylab.scatter(Y[:, 0], Y[:, 1], 20, labels)
            pylab.savefig(os.path.join('tsne_gif','iter{}.png'.format(iter)))
    plot_similarity(P,Q)
    
    # Return solution
    return Y
    
def ssne(X=np.array([]), no_dims=2, initial_dims=50, perplexity=30.0):
    """
        Runs t-SNE on the dataset in the NxD array X to reduce its
        dimensionality to no_dims dimensions. The syntaxis of the function is
        `Y = tsne.tsne(X, no_dims, perplexity), where X is an NxD NumPy array.
    """

    # Check inputs
    if isinstance(no_dims, float):
        print("Error: array X should have type float.")
        return -1
    if round(no_dims) != no_dims:
        print("Error: number of dimensions should be an integer.")
        return -1

    # Initialize variables
    X = pca(X, initial_dims).real
    (n, d) = X.shape
    max_iter = 1000
    initial_momentum = 0.5
    final_momentum = 0.8
    eta = 500
    min_gain = 0.01
    Y = np.random.randn(n, no_dims)
    dY = np.zeros((n, no_dims))
    iY = np.zeros((n, no_dims))
    gains = np.ones((n, no_dims))

    # Compute P-values
    P = x2p(X, 1e-5, perplexity)
    P = P + np.transpose(P)
    P = P / np.sum(P)
    P = P * 4.									# early exaggeration
    P = np.maximum(P, 1e-12)

    # Run iterations
    for iter in range(max_iter):

        # Compute pairwise affinities
        sum_Y = np.sum(np.square(Y), 1)
        num = -2. * np.dot(Y, Y.T)
        # modify 1
        num = 1  / (1 + num + sum_Y + sum_Y.reshape(-1,1))
        num[range(n), range(n)] = 0.
        Q = num / np.sum(num)
        Q = np.maximum(Q, 1e-12)

        # Compute gradient
        PQ = P - Q
        for i in range(n):
            # modify 2
            dY[i, :] = np.dot(PQ[i,:], Y[i,:] - Y)

        # Perform the update
        if iter < 20:
            momentum = initial_momentum
        else:
            momentum = final_momentum
        gains = (gains + 0.2) * ((dY > 0.) != (iY > 0.)) + \
                (gains * 0.8) * ((dY > 0.) == (iY > 0.))
        gains[gains < min_gain] = min_gain
        iY = momentum * iY - eta * (gains * dY)
        Y = Y + iY
        Y = Y - np.tile(np.mean(Y, 0), (n, 1))

        # Compute current value of cost function
        if (iter + 1) % 10 == 0:
            C = np.sum(P * np.log(P / Q))
            print("Iteration %d: error is %f" % (iter + 1, C))

        # Stop lying about P-values
        if iter == 100:
            P = P / 4.
    
    # save scatter plot
        if iter % 10 == 0:
            pylab.clf()
            pylab.scatter(Y[:, 0], Y[:, 1], 20, labels)
            pylab.savefig(os.path.join('ssne_gif','iter{}.png'.format(iter)))
    plot_similarity(P,Q)
    
    # Return solution
    return Y
    
    
def plot_similarity(P,Q):
    pylab.subplot(2,1,1)
    pylab.title('SSNE high-dim')
    pylab.hist(P.flatten(),bins=40,log=True)
    pylab.subplot(2,1,2)
    pylab.title('SSNE low-dim')
    pylab.hist(Q.flatten(),bins=40,log=True)
    pylab.show()



    
if __name__ == "__main__":
    print("Run Y = tsne.tsne(X, no_dims, perplexity) to perform t-SNE on your dataset.")
    print("Running example on 2,500 MNIST digits...")
    X = np.loadtxt("mnist2500_X.txt")
    labels = np.loadtxt("mnist2500_labels.txt")
    Y = tsne(X, 2, 50, 20.0)
    pylab.scatter(Y[:, 0], Y[:, 1], 20, labels)
    pylab.show()
    
    Y = ssne(X, 2, 50, 20.0)
    pylab.scatter(Y[:, 0], Y[:, 1], 20, labels)
    pylab.show()
```
